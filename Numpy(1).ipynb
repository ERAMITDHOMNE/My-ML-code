```python
# color image RGB  red color pixel , blue color value , green pixel
```

```python
broadcasting refer to the ability of numpy to 
treat array of different dimensionn/shape during performing arithmetic operation 
```

```python
import numpy as np
x2 = np.array([1,2,3]) # one dimension , because it is having only 1 row  , 1x3
x3 = np.array([[4,5,6],[7,8,9]]) # it is 2 dimension , it is having row and column ,
```

```python
x2 # 1x3
```

```python
x3 # 2 x 3
```

```python
print(x2 * x3)
```

# vectorization

```python
Numpy vectorization meaning performing mathematical operation on entire arrays,eliminating the need to loop through individual elements
It helps you to perform operation parallely on the array.
parallel running two or more operation same time so as to speedup the operation.

1 2 3
4 5 7

1x4
2x5
3x7

1x4 2x5 3x7
```
